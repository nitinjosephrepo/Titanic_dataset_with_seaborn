# Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_style('darkgrid')
```

# Load Dataset (Quick Access in Seaborn)

```python
#Load Dataset : Seaborn Provides quick access to a small number of example datasets 
titanic = sns.load_dataset('titanic')
```

# Sample Dataset  

```python
titanic.sample(5)
```

# Passengers - Age, Gender, Decks & Fare

Swarmplot of Age Distribution

```python
fig, ax = plt.subplots(figsize=(10,6)) 
sns.swarmplot(x=titanic["age"].dropna(), data = titanic, size = 4,palette = "rocket_r")
ax.set_title("Passenger Age Distribution")
```

Visualize Age by Man, Woman and Child

```python
plt.figure(figsize=(10,6))
sns.stripplot(x = 'age', y = 'who', data=titanic,palette = "rocket_r",jitter=0.08,marker="D")
plt.title("Age Distribution by Man, Woman & Child")
plt.tight_layout()
```

```python
sns.set_style('darkgrid')
ax = sns.catplot(x="class", hue="who",data=titanic, kind="count",palette = "rocket_r", height=6, aspect=1.8)
plt.title("People Count by Class")
plt.tight_layout()
```

```python
sns.catplot(x = "deck",hue ="who", kind ="count", data = titanic, palette = "rocket_r", height=5, aspect=2 )
plt.title("People Count by Deck")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.histplot(data = titanic, x = "who", y = "class", hue = "deck", palette = "ch:.25" )
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.histplot(titanic, x="class", y="age",hue = "who",legend=True,palette = "Set2")
```

```python
sns.catplot(x = "deck",y ="fare",hue ="who", kind ="swarm", data = titanic,palette = "rocket_r",\
            height=6, aspect=1.6)
plt.title("Fare by Deck")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.violinplot(x="deck", y="fare",hue="sex",data=titanic,palette = "rocket_r")
plt.title("Avg. Fare by Deck & Gender")
sns.despine(offset=10, trim=True)
```

```python
g = sns.jointplot(x='age', y='fare', data=titanic,height=7)
g.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color="r", height=-.15, clip_on=False)
 
```

# Survival - By Age, Gender, Deck & Embark Point

Did not Survive = 0, Survived = 1

```python
#Did not Survive = 0, Survived = 1
sns.displot(x = 'age',data = titanic, bins=[10, 20, 30, 40, 50, 60, 70,80,90],\
            hue = 'survived',multiple="stack", height = 8, palette = "rocket_r",)
```

```python
ax = sns.catplot(x="embark_town",hue="who", col="survived",data=titanic, kind="count",height=6, aspect=1,\
palette = "rocket_r")
ax.fig.suptitle('Survival By Embark Town')
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.stripplot(x = 'survived', y = 'age', hue='pclass', data=titanic, palette = 'ch:.25' )
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6)) 
sns.set_style('darkgrid')
sns.stripplot(x = 'sex',y = 'age',hue = "alive",data = titanic, palette = "rocket_r", size = 4)
plt.xticks(rotation=60)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5)) 
sns.countplot(x="pclass",data=titanic,hue ="survived",palette = "rocket_r")
ax.set_title("Total Number by Class")
```

```python
sns.set_style('darkgrid')
sns.catplot(x="class", hue="who",col = "survived", data=titanic, kind="count",palette = "rocket_r")
```

```python
sns.catplot(x="sex", y="survived",hue = "class", data=titanic, kind="bar",palette = "rocket_r",height=6, aspect=1.6)
plt.show()
```

```python
sns.catplot(x="sex",y = "age", hue="embark_town", col="survived",data=titanic,palette = "rocket_r", kind="strip",height=7, aspect=.7)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.violinplot(x="pclass", y="fare",hue="sex",data=titanic,palette = "rocket_r")
sns.despine(offset=10, trim=True)
```
